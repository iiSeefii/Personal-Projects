```python
data.insert(0, 'ID',[f'ID{i:03d}' for i in range(1, len(data)+1)]) 

#------------------------------------------------------------

data['INDICATOR'] = data['INDICATOR'].replace('Drug overdose death rates','DODR')

data['PANEL'] = data['PANEL'].replace('All drug overdose deaths','ADOD')
data['PANEL'] = data['PANEL'].replace('Drug overdose deaths involving any opioid','DODIO')
data['PANEL'] = data['PANEL'].replace('Drug overdose deaths involving heroin','DODIH')
data['PANEL'] = data['PANEL'].replace('Drug overdose deaths involving methadone','DODIM')
data['PANEL'] = data['PANEL'].replace('Drug overdose deaths involving natural and semisynthetic opioids','DODINSO')
data['PANEL'] = data['PANEL'].replace('Drug overdose deaths involving other synthetic opioids (other than methadone)','DODISO ! OTM')

data['STUB_NAME'] = data['STUB_NAME'].str.replace('and','&',regex = False)
data['STUB_NAME'] = data['STUB_NAME'].str.replace('race','Race',regex = False)
data['STUB_NAME'] = data['STUB_NAME'].str.replace('age','Age',regex = False)
data['STUB_NAME'] = data['STUB_NAME'].str.replace('Hispanic origin','H-Origin',regex = False)
data['STUB_NAME'] = data['STUB_NAME'].str.replace('(single Race)','(Single Race)',regex = False)

data['FLAG'] = data['FLAG'].replace('*','Unstable')
data['FLAG'] = data['FLAG'].fillna('No Notice')

data['PANEL_NUM'] = data['PANEL_NUM']+1

data['STUB_NAME_NUM'] = data['STUB_NAME_NUM']+1

data['UNIT'] = data['UNIT'].replace('Deaths per 100,000 resident population, crude','DP100K/RP/CRD')
data['UNIT'] = data['UNIT'].replace('Deaths per 100,000 resident population, age-adjusted','DP100K/RP/AGE-ADJ')

#------------------------------------------------------------

data['GENDER'] = data['STUB_LABEL'].str.extract('(Male|Female|All persons)' , expand = False)
data['RACE'] = data['STUB_LABEL'].str.extract('(White|Black or African American|Hispanic or Latino: All races|American Indian or Alaska Native|Not Hispanic or Latino: Black|Not Hispanic or Latino: White|Not Hispanic or Latino: American Indian or Alaska Native|Asian or Pacific Islander|Not Hispanic or Latino: Asian or Pacific Islander|Not Hispanic or Latino: Asian|Not Hispanic or Latino: Native Hawaiian or Other Pacific Islander)' , expand = False)

data['RACE'] = data['RACE'].str.replace('or','OR',regex = False)
data['RACE'] = data['RACE'].str.replace('African','Afro',regex = False)
data['RACE'] = data['RACE'].str.replace('Indian','Indi',regex = False)
data['RACE'] = data['RACE'].str.replace('Native','Nat',regex = False)
data['RACE'] = data['RACE'].str.replace('Hispanic','HIS',regex = False)
data['RACE'] = data['RACE'].str.replace('Latino','LAT',regex = False)
data['RACE'] = data['RACE'].str.replace('Hawaiian','HAWA',regex = False)
data['RACE'] = data['RACE'].str.replace('Pacific Islander','P-I',regex = False)
data['RACE'] = data['RACE'].str.replace('Other','~',regex = False)
data['RACE'] = data['RACE'].str.replace('Not HIS OR LAT:','',regex = False)

data['AGE-GROUP'] = data['STUB_LABEL'].str.extract('(Under 15|15-24|25-34|35-44|45-54|55-64|65-74|75-84|85 years and over)' , expand = False)

data['AGE-GROUP'] = data['AGE-GROUP'].replace('Under 15','Children')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('15-24','Youth')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('25-34','Young Adults')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('35-44','Adults')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('45-54','Middle-Aged')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('55-64','Pre-Seniors')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('65-74','Young Seniors')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('75-84','Seniors')
data['AGE-GROUP'] = data['AGE-GROUP'].replace('85 years and over','Oldest Seniors')
#------------------------------------------------------------

data['GENDER'] = data['GENDER'].replace('All persons','Any')

data['AGE-GROUP'] = data['AGE-GROUP'].fillna('Undefined')
data['ESTIMATE'] = data['ESTIMATE'].fillna('Undefined')
data['RACE'] = data['RACE'].fillna('Undefined')
data['GENDER'] = data['GENDER'].fillna('Any')


data['AGE_NUM'] = data['AGE_NUM'].replace({1.10:1,1.20:2,1.30:3,1.40:4,1.50:5,1.60:6,1.70:7,1.80:8,1.90:9,1.91:10}).astype(int)

#------------------------------------------------------------

data = data.drop(columns = ['STUB_LABEL','STUB_LABEL_NUM'])
```
